```python
# Dropping 'State' as it is irrelevent
train.drop(columns=['State'],inplace=True)
test.drop(columns=['State'],inplace=True)
```

```python
# One-Hot Encoding
train=pd.get_dummies(train,columns=['International plan','Voice mail plan','Churn'],drop_first=True)
test=pd.get_dummies(test,columns=['International plan','Voice mail plan','Churn'],drop_first=True)
```

```python
from re import X
from imblearn.over_sampling import SMOTE
from collections import Counter
x=train.drop(columns=['Churn_True'])
y=train['Churn_True']
sm = SMOTE(random_state=42)
print('Original dataset shape %s' % Counter(y))
X_train, y_train = sm.fit_resample(x, y)
print('Resampled dataset shape %s' % Counter(y_train))
```

```python
X_test=test.drop(columns=['Churn_True'])
y_test=test['Churn_True']
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import auc
from sklearn.metrics import precision_recall_curve
```

# Baseline DNN architecture without dropout layers

```python
# Baseline DNN architecture without dropout layers
model_baseline = Sequential([
    Dense(64, activation='relu', input_shape=(18,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model_baseline.compile(optimizer=Adam(), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])

# Train the model on the training set
model_baseline.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
earlystopper = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', min_delta=0, patience=15, verbose=1,mode='auto', baseline=None, restore_best_weights=False)
```

```python
history = model_baseline.fit(X_train, y_train, epochs=6, batch_size=5, validation_split=0.15, verbose=0, callbacks=[earlystopper])
history_dict = history.history
```

```python
loss_values = history_dict['loss']
val_loss_values=history_dict['val_loss']
plt.plot(loss_values,'b',label='training loss')
plt.plot(val_loss_values,'r',label='val training loss')
plt.legend()
plt.xlabel("Epochs")
```

```python
accuracy_values = history_dict['accuracy']
val_accuracy_values=history_dict['val_accuracy']
plt.plot(val_accuracy_values,'-r',label='val_accuracy')
plt.plot(accuracy_values,'-b',label='accuracy')
plt.legend()
plt.xlabel("Epochs")
```

```python
#predictions
baseline_predictions = model_baseline.predict(X_test)
```

```python
y_pred_binary = (baseline_predictions > 0.5).astype(int)
```

```python
#scores
print("Accuracy Neural Net:",metrics.accuracy_score(y_test, y_pred_binary))
print("Precision Neural Net:",metrics.precision_score(y_test, y_pred_binary))
print("Recall Neural Net:",metrics.recall_score(y_test, y_pred_binary))
print("F1 Score Neural Net:",metrics.f1_score(y_test, y_pred_binary))
```

```python
#CM matrix
matrix_nn = confusion_matrix(y_test, y_pred_binary)
cm_nn = pd.DataFrame(matrix_nn, index=['not_churn', 'churn'], columns=['not_churn', 'churn'])

sns.heatmap(cm_nn, annot=True, cbar=None, cmap="Blues", fmt = 'g')
plt.title("Confusion Matrix Neural Network"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

```python
# Predicted probabilities
y_pred_nn_proba = model_baseline.predict(X_test)
# Compute ROC curve and ROC-AUC score
fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test, y_pred_nn_proba)
auc_keras = auc(fpr_keras, tpr_keras)

# Print AUC score
print('AUC Neural Net:', auc_keras)
```

```python
#ROC
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras, tpr_keras, label='Keras (area = {:.3f})'.format(auc_keras))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Neural Net ROC curve')
plt.legend(loc='best')
plt.show()
```

```python
nn_precision, nn_recall, _ = precision_recall_curve(y_test, y_pred_nn_proba)
no_skill = len(y_test[y_test==1]) / len(y_test)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', color='black', label='No Skill')
plt.plot(nn_recall, nn_precision, color='orange', label='TF NN')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall curve')
plt.legend()
plt.show()
```

#DNN architecture with Monte Carlo Dropout

```python
# Extended DNN architecture with Monte Carlo Dropout
model_mc_dropout = Sequential([
    Dense(64, activation='relu', input_shape=(18,)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.4),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model_mc_dropout.compile(optimizer=Adam(), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])

# Train the model on the training set
model_mc_dropout.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
earlystopper = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', min_delta=0, patience=15, verbose=1,mode='auto', baseline=None, restore_best_weights=False)
```

```python
history3 = model_with_layerwise_dropout.fit(X_train, y_train, epochs=6, batch_size=5, validation_split=0.15, verbose=0, callbacks=[earlystopper])
history_dict3 = history3.history
```

```python
loss_values3 = history_dict3['loss']
val_loss_values3=history_dict3['val_loss']
plt.plot(loss_values3,'b',label='training loss')
plt.plot(val_loss_values3,'r',label='val training loss')
plt.legend()
plt.xlabel("Epochs")
```

```python
accuracy_values3 = history_dict3['accuracy']
val_accuracy_values3=history_dict3['val_accuracy']
plt.plot(val_accuracy_values3,'-r',label='val_accuracy')
plt.plot(accuracy_values3,'-b',label='accuracy')
plt.legend()
plt.xlabel("Epochs")
```

```python
# Evaluate the model with Monte Carlo Dropout on the validation set
num_mc_samples = 30  # Number of Monte Carlo samples for inference
mc_dropout_predictions = []

for _ in range(num_mc_samples):
    mc_dropout_predictions.append(model_mc_dropout.predict(X_test))

# Average the predictions
average_mc_dropout_predictions = tf.reduce_mean(mc_dropout_predictions, axis=0)
```

```python
y_pred_binary_mc_dropout = (average_mc_dropout_predictions.numpy() > 0.5).astype(int)
```

```python
# Scores
print("Accuracy Neural Net (MC Dropout):", metrics.accuracy_score(y_test, y_pred_binary_mc_dropout))
print("Precision Neural Net (MC Dropout):", metrics.precision_score(y_test, y_pred_binary_mc_dropout))
print("Recall Neural Net (MC Dropout):", metrics.recall_score(y_test, y_pred_binary_mc_dropout))
print("F1 Score Neural Net (MC Dropout):", metrics.f1_score(y_test, y_pred_binary_mc_dropout))
```

```python
# Confusion Matrix
matrix_nn_mc_dropout = confusion_matrix(y_test, y_pred_binary_mc_dropout)
cm_nn_mc_dropout = pd.DataFrame(matrix_nn_mc_dropout, index=['not_churn', 'churn'], columns=['not_churn', 'churn'])

sns.heatmap(cm_nn_mc_dropout, annot=True, cbar=None, cmap="Blues", fmt='g')
plt.title("Confusion Matrix Neural Network with MC Dropout")
plt.ylabel("True Class")
plt.xlabel("Predicted Class")
plt.show()
```

```python
# Predicted probabilities
y_pred_nn_proba_mc_dropout = model_mc_dropout.predict(X_test)

# Compute ROC curve and ROC-AUC score
fpr_keras_mc_dropout, tpr_keras_mc_dropout, thresholds_keras_mc_dropout = roc_curve(y_test, y_pred_nn_proba_mc_dropout)
auc_keras_mc_dropout = auc(fpr_keras_mc_dropout, tpr_keras_mc_dropout)

# Print AUC score
print('AUC Neural Net (MC Dropout):', auc_keras_mc_dropout)
```

```python
# ROC
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras_mc_dropout, tpr_keras_mc_dropout, label='Keras MC Dropout (area = {:.3f})'.format(auc_keras_mc_dropout))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Neural Net ROC curve with MC Dropout')
plt.legend(loc='best')
plt.show()
```

```python
# Precision-Recall curve
nn_precision_mc_dropout, nn_recall_mc_dropout, _ = precision_recall_curve(y_test, y_pred_nn_proba_mc_dropout)
no_skill = len(y_test[y_test == 1]) / len(y_test)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', color='black', label='No Skill')
plt.plot(nn_recall_mc_dropout, nn_precision_mc_dropout, color='orange', label='TF NN with MC Dropout')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall curve with MC Dropout')
plt.legend()
plt.show()
```
